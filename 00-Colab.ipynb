#**What is Google Colab?** 
**Google Colab** is an *online notebook-like coding environment* that is well-suited for *machine learning and data analysis*. It comes equipped *with many Machine Learning libraries* and offers GPU usage. It is mainly used by *data scientists and ML engineers*.

image.png


# **How do I get started with Google Colab?**
The way is to go over to your Google Drive account. In the top left corner select “New”, then “More” in the drop-down panel, and then “Google Collaboratory”.
